```python
# Calculate average percentages based on caste
caste_avg = df.groupby('Caste')[['Class_X_Percentage', 'Class_XII_Percentage']].mean()

print("\nAverage Class 10 and Class 12 Percentage by Caste:")
print(caste_avg)
```

```python
# Calculate average percentages based on father's occupation
father_occ_avg = df.groupby('Father_occupation')[['Class_X_Percentage', 'Class_XII_Percentage']].mean()

# Calculate average percentages based on mother's occupation
mother_occ_avg = df.groupby('Mother_occupation')[['Class_X_Percentage', 'Class_XII_Percentage']].mean()

print("\nAverage Class 10 and Class 12 Percentage by Father's Occupation:")
print(father_occ_avg)

print("\nAverage Class 10 and Class 12 Percentage by Mother's Occupation:")
print(mother_occ_avg)
```

```python
performance_avg = df.groupby('Performance')[['Class_X_Percentage', 'Class_XII_Percentage']].mean()

print("\nAverage Class 10 and Class 12 Percentage by Performance Category:")
print(performance_avg)
```

```python
performance_avg = df.groupby('Performance')[['Class_X_Percentage', 'Class_XII_Percentage']].mean()

print("\nAverage Class 10 and Class 12 Percentage by Performance Category:")
print(performance_avg)
```

```python
def plot_avg_bar_chart(group_by_column, title):
    avg_scores = df.groupby(group_by_column)[['Class_X_Percentage', 'Class_XII_Percentage']].mean()

    avg_scores.plot(kind='bar', figsize=(8, 5), colormap='viridis')
    plt.xlabel(group_by_column)
    plt.ylabel('Average Percentage')
    plt.title(title)
    plt.xticks(rotation=45)
    plt.legend(['Class 10', 'Class 12'])
    plt.show()
```

```python

plot_avg_bar_chart('coaching', 'Average Class 10 & 12 Percentage by Coaching Attendance')
plot_avg_bar_chart('Gender', 'Average Class 10 & 12 Percentage by Gender')
plot_avg_bar_chart('Caste', 'Average Class 10 & 12 Percentage by Caste')
plot_avg_bar_chart('Father_occupation', 'Average Class 10 & 12 Percentage by Father\'s Occupation')
plot_avg_bar_chart('Mother_occupation', 'Average Class 10 & 12 Percentage by Mother\'s Occupation')
plot_avg_bar_chart('Performance', 'Average Class 10 & 12 Percentage by Performance Category')
```
